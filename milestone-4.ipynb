```python
import torch
from TTS.api import TTS
import glob
```

```python
print(TTS().list_models())
```

```python
def convert_french_to_audio(dir_path, output_dir):
    device = "cuda" if torch.cuda.is_available() else "cpu"
    tts = TTS(model_name="tts_models/fr/mai/tacotron2-DDC").to(device)
    for f in glob.glob(dir_path):
        with open(f, 'r') as fr:
            text = fr.read()
        fname = f.split('/')[-1].replace('.txt', '')
        tts.tts_to_file(text, file_path=output_dir + fname + ".wav")
```

```python
convert_french_to_audio('../AI-Project/Translated_Text/*', '../AI-Project/Translated_Audio/')
```
